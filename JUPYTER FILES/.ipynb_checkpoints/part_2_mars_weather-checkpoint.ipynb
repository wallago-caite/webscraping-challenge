### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create a Beautiful Soup Object
html = browser.html
soup = BeautifulSoup(html,'html.parser')

type(soup)
```

```python
#Manual inspection of all elements in  website
# class container = div class_"container py-4"
# class container = div class_"container-fluid py-5"
#header h1 class_"display-5 fw-bold"
#table class_ "table" tr td th

#find tags non-manually
tags = []

for tag in soup.find_all(True):
    tags.append(tag.name)

set(tags)
```

```python
table = soup.find_all("table", class_ ="table")
table
```

```python
# Extract all rows of data (from 11.2 class discussion)
data_rows = soup.find_all("tr", class_="data-row")
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Heres an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
#extracted from 11.2 discussing in class

# Create an empty list
data_rows_list = []
# Loop through the scraped data to create a list of rows

for x in data_rows: #reminder that data_rows created is the dirty whole file
    row_values = x.find_all("td")
    value_list = []
    for x1 in row_values: # for loop within a for loop
        value_list.append(x1.text) #appending values to the inner list
    data_rows_list.append(value_list) #list within a list, appending list values to this new list

data_rows_list
```

```python
# find headers
table = soup.find('table')

headers = []

for th in table.find_all('th'):
    headers.append(th.get_text(strip=True))

print(headers)
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
df = pd.DataFrame(data_rows_list, columns= headers)
```

```python
# Confirm DataFrame was created successfully
display(df)
```

### Step 4: Prepare Data for Analysis

Examine the data types that are currently associated with each column. If necessary, cast (or convert) the data to the appropriate `datetime`, `int`, or `float` data types.

  > **Hint** You can use the Pandas `astype` and `to_datetime` methods to accomplish this task.


```python
# Examine data type of each column
df.dtypes
```

```python
# Change data types for data analysis
for column in df.columns:
    if column == "terrestrial_date":
        df[column] = pd.to_datetime(df[column])
    else:
        df[column] = pd.to_numeric(df[column])
```

```python
# Confirm type changes were successful by examining data types again
df.dtypes
```

### Step 5: Analyze the Data

Analyze your dataset by using Pandas functions to answer the following questions:

1. How many months exist on Mars?
2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?
3. What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
    * Find the average the minimum daily temperature for all of the months.
    * Plot the results as a bar chart.
4. Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
    * Find the average the daily atmospheric pressure of all the months.
    * Plot the results as a bar chart.
5. About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
    * Consider how many days elapse on Earth in the time that Mars circles the Sun once.
    * Visually estimate the result by plotting the daily minimum temperature.


```python
# 1. How many months are there on Mars?
months_on_mars = df['month'].max()

print(f"There are {months_on_mars} months on Mars.")
```

```python
# 2. How many Martian days' worth of data are there?
#df['sol'].max() # incorrect becauset this is the # of days since the lander landed 
martian_days_data = df['sol'].count() # this is the number of days in the DATASET itself

print(f"There are {martian_days_data} Martian days worth of data.")
```

```python
# 3. What is the average low temperature by month?
avg_low_temp = df["min_temp"].groupby(df['month']).mean()
avg_low_temp 

pd.DataFrame(avg_low_temp)

df_low_temp = pd.DataFrame(avg_low_temp)

df_low_temp
```

```python
# Plot the average temperature by month
avg_low_temp.plot.bar(rot = 0)
plt.ylabel("Temperature in Celcius")
plt.title("Average Low Temperature on Mars by Month")
plt.xlabel("Martian Month")
```

```python
# Identify the coldest and hottest months in Curiosity's location
avg_low_temp.sort_values().plot.bar(rot = 0)
plt.ylabel("Temperature in Celcius")
plt.title("Average Low Temperature on Mars by Month")
plt.xlabel("Martian Month")

coldest_min_temp = df_low_temp['min_temp'].min()
hottest_min_temp = df_low_temp['min_temp'].max()

coldest_month = df_low_temp[df_low_temp['min_temp'] == coldest_min_temp].index[0]
hottest_month = df_low_temp[df_low_temp['min_temp'] == hottest_min_temp].index[0]

print(f"The coldest average temperature was {coldest_min_temp:.2f} in Martian month {coldest_month}.")
print(f"The hottest average temperature was {hottest_min_temp:.2f} in Martian month {hottest_month}.")

```

```python
# 4. Average pressure by Martian month
avg_pressure = df["pressure"].groupby(df['month']).mean()
avg_pressure

pd.DataFrame(avg_pressure)

df_pressure = pd.DataFrame(avg_pressure)

df_pressure
```

```python
# Plot the average pressure by month
avg_pressure.sort_values().plot.bar(rot = 0)
plt.ylabel("Average Pressure")
plt.title("Average Atmospheric Pressure at Curiosity's Landsite")
plt.xlabel("Martian Month")

lowest_pressure = df_pressure['pressure'].min()
highest_pressure = df_pressure['pressure'].max()

low_pressure_month = df_pressure[df_pressure['pressure'] == lowest_pressure].index[0]
high_pressure_month = df_pressure[df_pressure['pressure'] == highest_pressure].index[0]

print(f"The highest pressure was {highest_pressure:.2f} in Martian month {high_pressure_month}.")
print(f"The lowest pressure was {lowest_pressure:.2f} in Martian month {low_pressure_month}.")
```

```python
# 5. How many terrestrial (earth) days are there in a Martian year?
# How many Martian days in the dataset-- find the max sol day
# How many Earth days are in the dataset-- find the total days from max to min 
min_earth_days = df['terrestrial_date'].min()
max_earth_days = df['terrestrial_date'].max()
earth_days_total =  max_earth_days - min_earth_days
earth_days_total.days
actual_earth_days = earth_days_total.days +1
actual_earth_days
```

```python
# Define a function to identify peaks
def is_peak(index, temp_data, threshold=0):
    if index == 0 or index == len(temp_data) - 1:
        return False
    if temp_data[index] > threshold and temp_data[index] > temp_data[index - 1] and temp_data[index] > temp_data[index + 1]:
        return True
    return False

# Find peaks using the function
peaks = []
for i in range(len(df["min_temp"])):
    if is_peak(i, df["min_temp"], threshold= -65):  # Adjust the threshold to -65
        peaks.append((i, df["min_temp"][i]))

peaks
```

On average, the third month has the coldest minimum temperature on Mars, and the eighth month is the warmest. But it is always very cold there in human terms!



Atmospheric pressure is, on average, lowest in the sixth month and highest in the ninth.

The distance from peak to peak is roughly 1425-750, or 675 days. A year on Mars appears to be about 675 days from the plot. Internet search confirms that a Mars year is equivalent to 687 earth days.
